```python
import pandas as pd
import seaborn as sns
import os
from os import listdir
import os.path as op
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'  # makes figs nicer!
```

```python
files = [os.path.join(dp, f) for dp, dn, fn in os.walk(os.path.expanduser("results")) for f in fn]
results_files = [f for f in files if ".csv" in f]
```

```python
tmp = []
for f in results_files:
    df_tmp = pd.read_csv(f)
    tmp.append(df_tmp)
```

```python
df_concat = pd.concat(tmp)
```

```python
df_concat['dataset'].value_counts()
```

```python
sns.catplot(data = df_concat, x = "model", y = "probability", 
            hue = "match", col = "dataset",kind = "bar")
```
